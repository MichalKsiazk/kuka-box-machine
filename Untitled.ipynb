```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def obrót(kąt, kąt2, dx, dy):
    theta = np.deg2rad(kąt)
    theta2 = np.deg2rad(kąt2)
    
    R1 = np.array([[np.cos(theta), -np.sin(theta), 0],
                  [np.sin(theta),  np.cos(theta),0],
                  [0,  0, 1]
                 ])
    R2 = np.array([[np.cos(theta2), -np.sin(theta2), 0],
                  [np.sin(theta2),  np.cos(theta2),0],
                  [0,  0, 1]
                 ])
    
    p1 = np.array([[1], [1], [1]])
    p2 = np.array([[1], [0], [0]])


    p1 = R1 @ p1
    
    p2 = R2 @ p2
    p3 = p1 + p2

    #p3 = p2.copy()
    #p3[2] = 1
    #p2 = p2[:2] - p1[:2]

   # p_t2 = T2 @ p_rot

    plt.figure(figsize=(5, 5))
    plt.arrow(0, 0, p1[0, 0], p1[1, 0], color='red', label='p1')
    plt.arrow(p1[0, 0], p1[1, 0], p3[0, 0], p3[1, 0], color='blue', label='p2')
    plt.plot(p3[0, 0], p3[1, 0], "o")
    #plt.arrow(p1[0, 0], p1[1, 0], p3[0, 0], p3[1, 0], color='green', label='p3')
    
    plt.xlim(-5, 5)
    plt.ylim(-5, 5)
    plt.grid(True)
    plt.gca().set_aspect('equal')
    plt.legend()
    plt.title(f"Obrót o {kąt}°")
    print("p1: ", p1[:2])
    print("p2: ", p2[:2])
    print("p3: ", p3[:2])
    plt.show()

interact(obrót, kąt=FloatSlider(min=0, max=360, step=1, value=0),
         kąt2=FloatSlider(min=0, max=360, step=1, value=0),
         dx=FloatSlider(min=-5, max=5, step=0.1, value=0),
         dy=FloatSlider(min=-5, max=5, step=0.1, value=0))
```
